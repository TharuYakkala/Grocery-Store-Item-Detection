```python
import torch
import torch.nn as nn
import torchvision.models as models
import numpy as np
```

```python
## Set device and check if cuda is available
device = "cuda" if torch.cuda.is_available() else 'cpu'
print(device)
```

```python
from torch_trainers.training_loop import main_trainer
from torch_trainers.torch_data_prepper import generate_dataloaders, augmented_transform
from torch_trainers.torch_custom_models import *
from torch.optim import Adam
```

```python
model, transform = make_mobilenet3_small(device=device)
```

```python
#model = make_vgg16(device=device)
optimizer = Adam(params=model.parameters(), lr=0.0001, weight_decay=0.0005)
loss_fn = nn.CrossEntropyLoss()
```

```python
train_dataloader, test_dataloader = generate_dataloaders(
    train_path="./data_augmented/train",
    test_path='./dataset/test_clean',
    transform=transform,
    batch_size=256,
    num_workers=4
)
```

```python
main_trainer(model=model,
             train_dataloader=train_dataloader,
             test_dataloader=test_dataloader,
             optimizer=optimizer,
             loss_fn=loss_fn,
             total_epochs=10,
             device=device)
```
